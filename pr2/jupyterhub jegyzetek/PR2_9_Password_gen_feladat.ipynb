## Véletlen jelszó generáló alkalmazás

https://passwordsgenerator.net/

<b>A program képes:</b>
<ol>
    <li>véletlen jelszó generálására</li>
    <li>felhasználó által megadott hosszúságú jelszó előállítására</li>
    <li>tetszőleges kombinációban megadott karakterkészletekből (kis-, nagybetűk, számok, speciális karakterek) generálni a jelszót</li>
    <li>ellenőrizni, hogy a generált jelszó megfelel-e a meghatározott feltételeknek</li>
    <li>mellékelni emlékeztető a jelszóhoz (fenti link, "Remember your password" szekció)</li>    
</ol>

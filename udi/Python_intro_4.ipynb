```python
import numpy as np
from matplotlib import pyplot as plt


def f(x):
   return np.sin(x)

def g(x):
   return np.cos(x)

def h(x):
   return 0.1 * x + 0.5

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='red')
plt.plot(x, g(x), color='blue')
plt.plot(x, h(x), color='green')

plt.show()
```

## 6. Feladat 

Van-e közös osztó?

```python
print("n: ", end="")
n=int(input())
print("m: ", end="")
m=int(input())
a=[]
b=[]
c=[]
for i in range (2,n+1):
  if n % i == 0:
   a.append(i)

print("\n")
for i in range (2,m+1):
  if m % i == 0:
   b.append(i)
print(a)
print(b)
n = len(b)
for elem in a:
  i=0
  while i<n and elem !=b[i]:
    i=i+1
  if i<n:
    c.append(elem)
print("közös osztók: ", c)
if c==[]:
  print("Nincs közös osztó")
else:
  print("Van közös osztó")
```
